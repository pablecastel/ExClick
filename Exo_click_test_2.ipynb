## 1. Objectives

- Select a specific country and study the impression, value and eCPM and its relationship.
- Observe specific time-driven patterns in these three metrics.

```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
# Read data
df = pd.read_csv('data_test2_clean.csv')
df.head()

# Convert datetime to datetime64
df['datetime'] = pd.to_datetime(df['datetime'])
```

# Investigate which country is interesting to dig in

```python
df_ecpm = df.groupby(['country']).agg({'eCPM': 'mean'}).reset_index()
df_ecpm.sort_values(by='eCPM', ascending=False, inplace=True)
df_ecpm.head()
```

```python
# Impressions per country
country_traffic = df.groupby(['country']).agg({'total_impressions': 'sum'}).reset_index()
country_traffic.sort_values(by='total_impressions', ascending=False, inplace=True)
country_traffic
```

```python
# Total value per country
country_value = df.groupby(['country']).agg({'total_value': 'sum'}).reset_index()
country_value.sort_values(by='total_value', ascending=False, inplace=True)
country_value
```

# Kenya 

Kenya is selected as specific country to analyze time trends of impressions, total_value and eCPM.


```python
# Filter by Kenya
df_kenya = df[df['country'] == 'KEN']

# Get day
df_kenya['day'] = df_kenya['datetime'].dt.day

# Range of time
start_date = df_kenya['datetime'].min()
end_date = df_kenya['datetime'].max()
print(f"Start date: {start_date}, End date: {end_date}")
```


# 1. Daily analysis


```python
# Group by day and sum total value and impressions
df_kenya_daily = df_kenya.groupby('day').agg({
    'total_value': 'sum',
    'total_impressions': 'sum'
}).reset_index()

# Compute eCPM by day
df_kenya_daily['eCPM'] = df_kenya_daily['total_value'] / df_kenya_daily['total_impressions'] * 1000

# Create 2-row subplot
fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    subplot_titles=('Total Value per Day', 'Total Impressions per Day'),
    vertical_spacing=0.04  # decrease (default is 0.08)
)

# Add total value line
fig.add_trace(
    go.Scatter(
        x=df_kenya_daily['day'], 
        y=df_kenya_daily['total_value'],
        mode='lines+markers',
        name='Total Value'
    ),
    row=1, col=1
)

# Add total impressions line
fig.add_trace(
    go.Scatter(
        x=df_kenya_daily['day'], 
        y=df_kenya_daily['total_impressions'],
        mode='lines+markers',
        name='Total Impressions'
    ),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        x=df_kenya_daily['day'], 
        y=df_kenya_daily['eCPM'],
        mode='lines+markers',
        name='eCPM'
    ),
    row=3, col=1
)

fig.update_layout(
    height=900, width=1400, 
    showlegend=False,
    xaxis_title='Day',
    xaxis2_title='Day',
    yaxis_title='Total Value',
    yaxis2_title='Total Impressions'
)

fig.show()
```

# 2. Hourly analysis

```python
# Group by datetime
df_kenya_hourly = df_kenya.groupby('datetime').agg({
    'total_value': 'sum',
    'total_impressions': 'sum'
}).reset_index()

# Compute eCPM
df_kenya_hourly['eCPM'] = df_kenya_hourly['total_value'] / df_kenya_hourly['total_impressions'] * 1000

# Create 3-row subplot, reduce distance between subplots with 'vertical_spacing'
fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    subplot_titles=('Total Value per Day', 'Total Impressions per Day'),
    vertical_spacing=0.04  # decrease (default is 0.08)
)

# Add total value line (no markers)
fig.add_trace(
    go.Scatter(
        x=df_kenya_hourly['datetime'], 
        y=df_kenya_hourly['total_value'],
        mode='lines',
        name='Total Value'
    ),
    row=1, col=1
)

# Add total impressions line (no markers)
fig.add_trace(
    go.Scatter(
        x=df_kenya_hourly['datetime'], 
        y=df_kenya_hourly['total_impressions'],
        mode='lines',
        name='Total Impressions'
    ),
    row=2, col=1
)

# Add eCPM line (no markers)
fig.add_trace(
    go.Scatter(
        x=df_kenya_hourly['datetime'], 
        y=df_kenya_hourly['eCPM'],
        mode='lines',
        name='eCPM'
    ),
    row=3, col=1
)

fig.update_layout(
    height=900, width=1400, 
    showlegend=False,
    xaxis_title='Day',
    xaxis2_title='Day',
    yaxis_title='Total Value',
    yaxis2_title='Total Impressions'
)

fig.show()
```

```python
# Compute day of the week
df_kenya['day_of_week'] = df_kenya['datetime'].dt.dayofweek

# Map day of the week to name
df_kenya['day_of_week'] = df_kenya['day_of_week'].map({
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
})

# Get hour of the day
df_kenya['hour'] = df_kenya['datetime'].dt.hour

df_kenya
```

```python
# Compute mean and 95% confidence interval per day of the week and hour
import scipy.stats as st

def ci_95(series):
    n = series.count()
    if n < 2:
        return float('nan')
    std = series.std()
    se = std / n**0.5
    h = se * st.t.ppf(1 - 0.025, n - 1)
    return h

df_kenya_day_of_week = df_kenya.groupby(['day_of_week', 'hour']).agg(
    {
        'total_value': ['mean', ci_95],
        'total_impressions': ['mean', ci_95],
        'eCPM': ['mean', ci_95]
    }
).reset_index()

# Optionally flatten MultiIndex columns and rename CI columns for clarity
df_kenya_day_of_week.columns = [
    '_'.join(col).replace('ci_95', 'ci95').strip('_') for col in df_kenya_day_of_week.columns.values
]
df_kenya_day_of_week
```

```python
import plotly.graph_objects as go

# Sort the DataFrame by natural day of week, then by hour
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plot_df = df_kenya_day_of_week.copy()
plot_df['day_of_week'] = pd.Categorical(plot_df['day_of_week'], categories=day_order, ordered=True)
plot_df = plot_df.sort_values(['day_of_week', 'hour'])

# For improved x-axis labeling, create a combined label
plot_df['day_hour'] = plot_df['day_of_week'].astype(str) + ' ' + plot_df['hour'].astype(str)

# Calculate bounds for confidence interval (mean +/- ci95)
plot_df['ci95_upper'] = plot_df['total_value_mean'] + plot_df['total_value_ci95']
plot_df['ci95_lower'] = plot_df['total_value_mean'] - plot_df['total_value_ci95']

fig = go.Figure()

# Add shaded area for mean  95% CI
fig.add_trace(
    go.Scatter(
        x=pd.concat([plot_df['day_hour'], plot_df['day_hour'][::-1]]),
        y=pd.concat([plot_df['ci95_upper'], plot_df['ci95_lower'][::-1]]),
        fill='toself',
        fillcolor='rgba(58, 114, 189, 0.2)',  # blueish, transparent
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=True,
        name='95% CI'
    )
)

# Plot the mean line
fig.add_trace(
    go.Scatter(
        x=plot_df['day_hour'],
        y=plot_df['total_value_mean'],
        mode='lines+markers',
        name='Mean Total Value',
        line=dict(color='red')
    )
)

fig.update_layout(
    title='Kenya: Total Value Mean and 95% CI by Day of Week and Hour',
    xaxis_title='Day of Week and Hour',
    yaxis_title='Total Value',
    legend_title='Metric',
    height=500,
    width=1400
)

fig.show()
```

# Discussion

## Kenya  Ad Performance Insights

### 1. Stable Monthly Performance

There is a clear mirrored pattern between total impressions and total value, which means revenue is mainly driven by volume.  
As impressions increase or decrease, total value follows the same trend.

eCPM remains stable throughout the month, indicating consistent advertiser demand and no strong monthly fluctuations.

---

### 2. No Best Day of the Month

There is no specific day of the month that stands out.  
Performance is relatively stable across the entire month, which makes forecasting easier and more predictable.

---

### 3. Clear Daily Pattern

While the month is stable, there is a clear hourly trend repeated throughout the week.

Ad performance peaks around:
- **07:00**
- **23:00**

These hours likely reflect higher user activity in the early morning and late evening.

---

### 4. Recommendation

The best times to display ads in Kenya are around:
- **06:0008:00**
- **22:0023:00**

Focusing on these hours can help maximize visibility and performance.
